### Extracting values with Regex
The heights and weights are stored as objects and are in imperial. They need converting into a useful format. I'll use a regular expression to extract all the feet and inch values, then convert them into cm.


```python
# Extracting the height and storing it in cm in the main table
ft_in=fifa['Height'].str.extract("(\d)'(\d\d?)").astype(int)
fifa['Height_cm']= (ft_in[0]*30.48+ft_in[1]*2.54).round(0)
```

```python
# Extracting the weight and storing it in kg
lbs=fifa['Weight'].str.extract("(\d+)lbs").astype(int)
fifa['Weight_kg'] = (lbs[0]*0.453592).round(1)
```

Fortunately, all the player wages and values are stored in Euros. This makes extracting and comparing the values much simpler. The values are either suffixed with M or K.<br>
All wages are suffixed with K or nothing (for players with wages under €1000).

```python
value_=fifa['Value'].str.extract("€(\d+\.?\d+?)M|€(\d+\.?\d+?)K").astype(float).fillna(0)
fifa['Value_']= value_[0]*10**6 + value_[1]*10**3
```

```python
wage_ = fifa['Wage'].str.extract("€(\d+\.?\d+?)K|€(\d+)").astype(float).fillna(0)
fifa['Wage_']=wage_[0]*1000 + wage_[1]
```
